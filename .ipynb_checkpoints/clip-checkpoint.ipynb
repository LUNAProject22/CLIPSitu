# Feature Engineering
Generate Clip embeddings for imSitu dataset -> concat(Image + Verb + Role)

```python
# Load the model
device = "cuda:2" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load('ViT-B/32', device)
```

```python
all_features = []
all_targets = []
img_dir = '/raid/data_vermad/of500_images_resized/'

with torch.no_grad():
    for img in train_images:
        img_path = img_dir + img
        img_input = preprocess(Image.open(img_path))
        img_features = model.encode_image(img_input.unsqueeze(0).to(device))
        verb = train[img]['verb']
        verb_features = model.encode_text(clip.tokenize(verb).to(device))
        roles = list(verbs[verb]['roles'].keys())
        for role in roles:
            role_features = model.encode_text(clip.tokenize(role).to(device))
            # choose one frame randomly
            selected_frame = random.choice(train[img]['frames'])
            # if target noun is missing then skip to next role
            if selected_frame[role]=='':
                break                   
            target = nouns[selected_frame[role]]['gloss'][0]
            
            cat_features = torch.cat((img_features,verb_features,role_features),1)
            target_features = model.encode_text(clip.tokenize(target).to(device))
            all_features.append(cat_features)
            all_targets.append(target)
```

```python
# feed in json dataset, either train.json or dev.json 
def get_features(dataset,target_vec=False):
    images = list(dataset.keys())
    img_dir = '/raid/data_vermad/imsitu_clip/of500_images_resized/'

    all_features = []
    all_targets = []
    with torch.no_grad():
        for img in tqdm(images):
            img_path = img_dir + img
            img_input = preprocess(Image.open(img_path))
            img_features = model.encode_image(img_input.unsqueeze(0).to(device))
            verb = dataset[img]['verb']
            verb_features = model.encode_text(clip.tokenize(verb).to(device))
            roles = list(verbs[verb]['roles'].keys())
            for role in roles:
                role_features = model.encode_text(clip.tokenize(role).to(device))
                # choose one frame randomly
                selected_frame = random.choice(dataset[img]['frames'])
                # if target noun is missing then skip to next role
                if selected_frame[role]=='':
                    break                   
                target = nouns[selected_frame[role]]['gloss'][0]
                
                cat_features = torch.cat((img_features,verb_features,role_features),1)
                all_features.append(cat_features)
                if(target_vec==True):
                    target_features = model.encode_text(clip.tokenize(target).to(device))
                    all_targets.append(target_features)
                else:
                    all_targets.append(target)

        if(target_vec==True):
            all_targets = torch.cat(all_targets)

    return torch.cat(all_features), all_targets
```

```python
def get_target_vector(targets):
    all_targets = []
    for target in targets:
        target_features = model.encode_text(clip.tokenize(target).to(device))
        all_targets.append(target_features)
    return torch.cat(all_targets)
```

# Modeling

```python
neural_net = nn.Sequential(
    nn.Linear(1536, 1024),
    nn.ReLU(),
    nn.Linear(1024, 512)
)
model = neural_net().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(neural_net.parameters(), lr=0.001, momentum=0.9)
```

```python
num_epochs = 1000
```

```python
%%time
# keeping-track-of-losses 
train_losses = []
valid_losses = []

for epoch in range(1, num_epochs + 1):
    # keep-track-of-training-and-validation-loss
    train_loss = 0.0
    valid_loss = 0.0
    
    # training-the-model
    model.train()
    for data, target in train_loader:
        # move-tensors-to-GPU 
        #data = data.to(device)
        #target = target.to(device)
        
        # clear-the-gradients-of-all-optimized-variables
        optimizer.zero_grad()
        # forward-pass: compute-predicted-outputs-by-passing-inputs-to-the-model
        output = model(data)
        # calculate-the-batch-loss
        loss = criterion(output, target)
        # backward-pass: compute-gradient-of-the-loss-wrt-model-parameters
        loss.backward()
        # perform-a-ingle-optimization-step (parameter-update)
        optimizer.step()
        # update-training-loss
        train_loss += loss.item() * data.size(0)
        
    # validate-the-model
    model.eval()
    for data, target in valid_loader:
        
        data = data.to(device)
        target = target.to(device)
        
        output = model(data)
        
        loss = criterion(output, target)
        
        # update-average-validation-loss 
        valid_loss += loss.item() * data.size(0)
    
    # calculate-average-losses
    train_loss = train_loss/len(train_loader.sampler)
    valid_loss = valid_loss/len(valid_loader.sampler)
    train_losses.append(train_loss)
    valid_losses.append(valid_loss)
        
    # print-training/validation-statistics 
    print('Epoch: {} \tTraining Loss: {:.6f} \tValidation Loss: {:.6f}'.format(
        epoch, train_loss, valid_loss))
```
