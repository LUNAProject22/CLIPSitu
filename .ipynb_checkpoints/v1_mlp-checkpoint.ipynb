# Feature Engineering
Generate Clip embeddings for imSitu dataset -> concat(Image + Verb + Role)

```python
# Load the model
device = "cuda:2" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load('ViT-B/32', device)
```

### Get features

```python
# feed in json dataset, either train.json, dev.json or test.json 
# returns dictionary of lists containing image filenames, verbs, roles, nouns, features_embeddings, target_embeddings 
def extract_data(dataset):
    images = list(dataset.keys())
    img_dir = '/raid/data_vermad/imsitu_clip/of500_images_resized/'
    all_images = []
    all_verbs = []
    all_roles = []
    all_nouns = []
    all_features = []
    all_targets = []
    with torch.no_grad():
        for img in tqdm(images):
            img_path = img_dir + img
            img_input = preprocess(Image.open(img_path))
            img_features = model.encode_image(img_input.unsqueeze(0).to(device))
            verb = dataset[img]['verb']
            verb_features = model.encode_text(clip.tokenize(verb).to(device))
            roles = list(verbs[verb]['roles'].keys())
            for frame in dataset[img]['frames']:
                for role in frame:
                    if frame[role]=='':
                        pass
                    else:
                        role_features = model.encode_text(clip.tokenize(role).to(device))
                        noun = nouns[frame[role]]['gloss'][0]
        
                        cat_features = torch.cat((img_features,verb_features,role_features),1)
                        target_features = model.encode_text(clip.tokenize(noun).to(device))
                        all_features.append(cat_features)
                        all_images.append(img)
                        all_verbs.append(verb)
                        all_roles.append(role)
                        all_nouns.append(noun)
                        all_targets.append(target_features)

        list_dict = {"images":all_images,"verbs":all_verbs,"roles":all_roles,"nouns":all_nouns,"embed_ft":all_features,"embed_target":all_targets}
    return list_dict
```

#### Get Target Embeddings

```python
def get_target_vector(targets):
    with torch.no_grad():
        all_targets = []
        for target in targets:
            target_feature = model.encode_text(clip.tokenize(target).to(device))
            all_targets.append(target_feature)
    return torch.cat(all_targets)
```

### Build Torch Dataset and Dataloaders

```python
class imSituEmbeddings(Dataset):
    def __init__(self, X, Y):
        self.X = X.type(torch.float32)
        self.y = Y.type(torch.float32)
        
    def __len__(self):
        return len(self.y)
    
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]
```

# Modeling

```python
neural_net = nn.Sequential(
    nn.Linear(1536, 1024),
    nn.ReLU(),
    nn.Linear(1024, 512)
).to(device)

criterion = nn.L1Loss()
optimizer = optim.SGD(neural_net.parameters(), lr=0.001, momentum=0.9)
```

```python
num_epochs = 500
```

```python
%%time
# keeping-track-of-losses 
train_losses = []
valid_losses = []

for epoch in range(1, num_epochs + 1):
    # keep-track-of-training-and-validation-loss
    train_loss = 0.0
    valid_loss = 0.0
    
    # training-the-model
    neural_net.train()
    for data, target in train_loader:
        # move-tensors-to-GPU 
        data = data.to(device)
        target = target.to(device)
        
        # clear-the-gradients-of-all-optimized-variables
        optimizer.zero_grad()
        # forward-pass: compute-predicted-outputs-by-passing-inputs-to-the-model
        output = neural_net(data)
        # calculate-the-batch-loss
        loss = criterion(output, target)
        # backward-pass: compute-gradient-of-the-loss-wrt-model-parameters
        loss.backward()
        # perform-a-ingle-optimization-step (parameter-update)
        optimizer.step()
        # update-training-loss
        train_loss += loss.item() * data.size(0)
        
    # validate-the-model
    neural_net.eval()
    for data, target in valid_loader:
        
        data = data.to(device)
        target = target.to(device)
        
        output = neural_net(data)
        
        loss = criterion(output, target)
        
        # update-average-validation-loss 
        valid_loss += loss.item() * data.size(0)
    
    # calculate-average-losses
    train_loss = train_loss/len(train_loader)
    valid_loss = valid_loss/len(valid_loader)
    train_losses.append(train_loss)
    valid_losses.append(valid_loss)
        
    # print-training/validation-statistics 
    print('Epoch: {} \tTraining Loss: {:.6f} \tValidation Loss: {:.6f}'.format(
        epoch, train_loss, valid_loss))
```

### Top K Predicted Labels

```python
predicted_topn_labels = []
for sample in all_predictions:
    distance = torch.norm(train_label_embeddings - sample, dim=1)
    nearest_k = torch.topk(distance,largest=False,dim=1).indices

    labels = [train_labels[x] for x in nearest_k]
    predicted_topn_labels.append(labels)
```

### Top Predicted Label

```python
predicted_labels = []
for sample in all_predictions:
    distance = torch.norm(train_label_embeddings - sample, dim=1)
    nearest = torch.argmin(distance)
    label = train_labels[nearest]
    predicted_labels.append(label)
```
