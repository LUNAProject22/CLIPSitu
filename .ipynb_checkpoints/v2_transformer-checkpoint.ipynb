```python
# Load the model
device = "cuda:2" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load('ViT-B/32', device)
```

```python
# feed in json dataset, either train.json, dev.json or test.json 
# returns dictionary of lists containing image filenames, verbs, roles, nouns, features_embeddings, target_embeddings 
def extract_data(dataset):
    images = list(dataset.keys())
    img_dir = '/raid/data_vermad/imsitu_clip/of500_images_resized/'
    all_frame_features = []
    all_targets = []
    encoded_features = []
    encoded_targets = []
    with torch.no_grad():
        for img in tqdm(images):
            img_path = img_dir + img
            img_input = preprocess(Image.open(img_path))
            img_features = model.encode_image(img_input.unsqueeze(0).to(device))
            verb = dataset[img]['verb']
            verb_features = model.encode_text(clip.tokenize(verb).to(device))
            #roles = list(verbs[verb]['roles'].keys())
            for frame in dataset[img]['frames']:
                framelist = []
                targetlist = []
                encoded_framelist = []
                encoded_targetlist = []
                for role in frame:
                    if frame[role]=='':
                        pass
                    else:
                        role_features = model.encode_text(clip.tokenize(role).to(device))
                        noun = nouns[frame[role]]['gloss'][0]
                        framelist.append([img,verb,role])
                        targetlist.append(noun)
                        cat_features = torch.squeeze(torch.cat((img_features,verb_features,role_features),1))
                        target_features = model.encode_text(clip.tokenize(noun).to(device))
                        encoded_framelist.append(cat_features)
                        encoded_targetlist.append(target_features)
                encoded_features.append(encoded_framelist)
                encoded_targets.append(encoded_targetlist)
                all_frame_features.append(framelist)
                all_targets.append(targetlist)


        list_dict = {"frames":all_frame_features,"targets":all_targets,"embed_ft":encoded_features,"embed_target":encoded_targets}
    return list_dict
```

```python
train_dict = extract_data(train)
```

```python
pickle.dump(train_dict, open('/raid/data_vermad/imsitu_clip/data/processed/train_frame_dict.pickle', 'wb'))
```

```python

class Transformer(nn.Module):
    def __init__(
        self,
        num_tokens,
        dim_model,
        num_heads,
        num_encoder_layers,
        num_decoder_layers,
        dropout_p,
    ):
        super().__init__()

        # Layers
        self.transformer = nn.Transformer(
            d_model=dim_model,
            nhead=num_heads,
            num_encoder_layers=num_encoder_layers,
            num_decoder_layers=num_decoder_layers,
            dropout=dropout_p,
        )

    def forward(self, src, target):
        src = src.permute(1, 0, 2)
        tgt = tgt.permute(1, 0, 2)

        # Transformer blocks - Out size = (sequence length, batch_size, num_tokens)
        transformer_out = self.transformer(src, tgt)
        out = self.out(transformer_out)
```
